```python
#15.1 Use multiprocessing to create three separate processes. 
#Make each one wait a random number of seconds between zero and one, 
#print the current time, and then exit.
import multiprocessing
import time
import random
import os
def theTime(what):
    time.sleep(random.random())
    now = time.time()
    print (" Process %s says: %s" % (os.getpid(),what))

if __name__ == "__main__":
    #theTime("im the main program")

    processes = []  # Create a list to keep track of all processes
    
    for n in range(3):
        now = time.time()
        p = multiprocessing.Process(target=theTime,    
            args=(time.ctime(now),)
        )
        processes.append(p) # Add to our list
        p.start()
    for p in processes:
        p.join()
    print("All processes completed")

# works on vs code and not on jupyternotes for jupternotes reasons
        
```
